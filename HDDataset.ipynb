```python
from ucimlrepo import fetch_ucirepo 
  
 
heart_disease = fetch_ucirepo(id=45) 
  
X = heart_disease.data.features 
y = heart_disease.data.targets 
  

print(heart_disease.metadata) 
print(heart_disease.variables) 
```

```python
import pandas as pd
```

```python
df = pd.concat([X, y], axis=1)

print(f"Final shape: {df.shape}")
print(df.head())
```

```python
df['ca_is_null'] = df['ca'].isnull()

# 2. Comparem 'ca_is_null' amb la columna 'num'
print("--- Relació entre Nuls a 'ca' i la columna 'num' ---")
print(pd.crosstab(df['ca_is_null'], df['num']))
```

```python
df['thal_is_null'] = df['thal'].isnull()
print("\n--- Relació entre Nuls a 'thal' i la columna 'num' ---")
print(pd.crosstab(df['thal_is_null'], df['num']))
```

```python
df.dropna(subset=['thal'], inplace=True)
```

```python
df.loc[(df['ca'].isnull()) & (df['num'] == 0), 'ca'] = 0.0

df['ca'] = df['ca'].fillna(1.0)

print("Nuls restants a 'ca':", df['ca'].isnull().sum())
print("Nuls totals al DF:", df.isnull().sum().sum())
```

```python
df.info()
```

```python
df['patient_id'] = range(1, len(df) + 1)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
heart_main = df.copy()
heart_main = heart_main.drop(columns=['ca_is_null', 'thal_is_null'], errors='ignore')

# 2. Movem el 'patient_id' al principi (posició 0)
# Primer el treiem i després l'inserim a la primera posició
cols = list(heart_main.columns)
if 'patient_id' in cols:
    cols.insert(0, cols.pop(cols.index('patient_id')))
    heart_main = heart_main[cols]

# 3. Renomenem les columnes categòriques a IDs
heart_main = heart_main.rename(columns={
    'sex': 'sex_id',
    'cp': 'cp_id',
    'fbs': 'fbs_id',
    'restecg': 'restecg_id',
    'slope': 'slope_id',
    'thal': 'thal_id',
    'num': 'diagnosis_id',
    'exang': 'exang_id'
})

print("Columnes actuals de heart_main:")
print(heart_main.columns)
```

```python
heart_main['thal_id'] = heart_main['thal_id'].replace({3: 1, 6: 2, 7: 3})
```

```python
heart_main.head()
```

```python
dim_sex = pd.DataFrame({
    'sex_id': [0, 1], 
    'sex_desc': ['Female', 'Male']
})

dim_cp = pd.DataFrame({
    'cp_id': [1, 2, 3, 4], 
    'cp_desc': ['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic']
})

dim_fbs = pd.DataFrame({
    'fbs_id': [0, 1], 
    'fbs_desc': ['<= 120 mg/dL', '> 120 mg/dL']
})

dim_restecg = pd.DataFrame({
    'restecg_id': [0, 1, 2], 
    'restecg_desc': ['Normal', 'ST-T abnormality', 'LV hypertrophy']
})

dim_exang = pd.DataFrame({
    'exang_id': [0, 1],
    'exang_desc': ['No exercise-induced angina', 'Exercise-induced angina'],
    'exang_risc': ['Low risk', 'High risk']
})

dim_slope = pd.DataFrame({
    'slope_id': [1, 2, 3], 
    'slope_desc': ['Upsloping', 'Flat', 'Downsloping']
})

dim_thal = pd.DataFrame({
    'thal_id': [1, 2, 3], 
    'thal_desc': ['Normal', 'Fixed defect', 'Reversible defect']
})

dim_target = pd.DataFrame({
    'diagnosis_id': [0, 1, 2, 3, 4], 
    'diagnosis_desc': ['Healthy', 'Stage 1', 'Stage 2', 'Stage 3', 'Stage 4']
})

print("All tables has been created correctly.")
```

```python
# --- FINAL STRUCTURE CHECK ---

# Display the main fact table
print("=== FACT TABLE: heart_main ===")
display(heart_main.head())

# Create a dictionary to iterate through all dimension tables
dimension_tables = {
    "Dimension: Sex": dim_sex,
    "Dimension: Chest Pain (CP)": dim_cp,
    "Dimension: Fasting Blood Sugar (FBS)": dim_fbs,
    "Dimension: Resting ECG": dim_restecg,
    "Dimension: Exercise Angina (Exang)": dim_exang,
    "Dimension: Slope": dim_slope,
    "Dimension: Thalassemia (Thal)": dim_thal,
    "Dimension: Diagnosis (Target)": dim_target
}

# Print headers and heads for each dimension table
for name, table in dimension_tables.items():
    print(f"\n=== {name} ===")
    display(table.head())
```

```python
# --- FINAL EXPORT TO CSV (CLEANED DATA) ---

# 1. Export the Fact Table (The main table with patient data)
# This includes the patient_id, clinical measurements, and foreign keys (_id)
heart_main.to_csv('heart_main.csv', index=False)

# 2. Export all Dimension Tables (The lookup dictionaries)
# Make sure you have executed the thal_id remap (1, 2, 3) before this!
dim_sex.to_csv('dim_sex.csv', index=False)
dim_cp.to_csv('dim_cp.csv', index=False)
dim_fbs.to_csv('dim_fbs.csv', index=False)
dim_restecg.to_csv('dim_restecg.csv', index=False)
dim_exang.to_csv('dim_exang.csv', index=False)
dim_slope.to_csv('dim_slope.csv', index=False)
dim_thal.to_csv('dim_thal.csv', index=False)
dim_target.to_csv('dim_target.csv', index=False)

print("🚀 SUCCESS: All 9 CSV files have been exported and are ready for SQL!")
```
